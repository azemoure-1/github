## Data Splitter

```python
import pandas as pd
df = pd.read_csv('contributors_df.csv')
```

```python
# Mélanger la trame de données à l'aide de la fonction
df = df.sample(frac=1)
df
```

```python

# Select ratio
ratio_1 = 0.2
ratio_2 = 0.2
ratio_3 = 0.6

total_rows = df.shape[0]

# Calculer le nombre de rangées pour chaque partie
part1 = int(total_rows * ratio_1)
part2 = int(total_rows * (ratio_1 + ratio_2))
part3 = total_rows

# Diviser les données
part_JSON = df.iloc[:part1]
part_CSV = df.iloc[part1:part2] 
part_df = df.iloc[part2:]
```
