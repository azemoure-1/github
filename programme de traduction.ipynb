```python
import pandas as pd
import langid
from translate import Translator

# Fonction de traduction de texte à l’aide de l’API Google Translate
def translate_text(text, target_language):
    if isinstance(text, str):
        translator = Translator(to_lang=target_language)
        translation = translator.translate(text)
        return translation
    else:
        return text

# Invite l’utilisateur à saisir
data_file = input("Enter the name of your data file (including the extension): ")
column_to_translate = input("Enter the column name to translate: ")
target_language = input("Enter the target language: ")

# Charger le fichier de données dans un DataFrame
df = pd.read_csv(data_file)

# Traduire la colonne spécifiée
df[column_to_translate] = df[column_to_translate].apply(lambda x: translate_text(x, target_language)
                                                       if isinstance(x, str) and langid.classify(x)[0] != target_language
                                                       else x)

print(df)
```
